```python
r = np.array([1.2,-0.1])
t = np.array([0.8,-0.8])

t_par_r = r * (np.dot(t,r) / np.dot(r,r))
print(t_par_r)

t_per_r = t - t_par_r
print(t_per_r)

plt.figure(figsize=(6,6))
plt.plot([0,t[0]],[0,t[1]],color='k',linewidth=3,label=r'')
plt.plot([0,r[0]],[0,r[1]],color=[.7,.7,.7],linewidth=3,label=r'')
plt.plot([0,t_par_r[0]],[0,t_par_r[1]],'k--',linewidth=3,label=r'')
plt.plot([0,t_per_r[0]],[0,t_per_r[1]],'k:',linewidth=3,label=r'')

print(t_par_r + t_per_r)
print(np.dot(t_par_r, t_per_r))
```

```python
l1 = 1
l2 = 2
l3 = -3
v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

l1*v1 + l2*v2 + l3*v3
```

```python
# Exercise 3-1
scalars = [l1,l2,l3,5]
vectors = [v1,v2,v3]

lw_comb = np.zeros(3)

for s,v in zip(scalars, vectors):
    lw_comb += s*v
lw_comb
```

```python
# Exercise 3-3
A = np.array([1,3])

xlim = [-4,4]

scalars = np.random.uniform(low=xlim[0], high=xlim[1], size = 100)

plt.figure()

for s in scalars:
    p = A*s
    plt.plot(p[0], p[1], "ko")
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

v1 = np.array([3,5,1])
v2 = np.array([0,2,2])

scalars = np.random.uniform(low=xlim[0], high=xlim[1], size = (100,2))

points = np.zeros((100,3))

for i in range(len(scalars)):
    points[i,:] = v1*scalars[i,0] + v2*scalars[i,1]

df = pd.DataFrame({
    'x_col': points[:,0],
    'y_col': points[:,1],
    'z_col': points[:,2]
})

fig = px.scatter_3d(
    df,
    x='x_col',
    y='y_col',
    z='z_col',
)

fig.show()
```
